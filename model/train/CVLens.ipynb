```python
def get_spacy_doc(error_file, data):
    nlp = spacy.blank('en')
    db = DocBin()
    count = 0

    # Each entry in data is a dictionary with key "annotations"
    for entry in tqdm(data):
        annotations_list = entry.get("annotations", [])
        # Process each annotation pair in the list
        for annotation_pair in annotations_list:
            # Skip if annotation_pair is None or not a list of length 2
            if not annotation_pair or not isinstance(annotation_pair, list) or len(annotation_pair) != 2:
                continue

            text, annot = annotation_pair
            doc = nlp.make_doc(text)
            annot = annot.get('entities', [])
            ents = []
            entity_indices = []

            for start, end, label in annot:
                # Skip overlapping entities
                if any(idx in entity_indices for idx in range(start, end)):
                    continue

                entity_indices.extend(range(start, end))
                span = doc.char_span(start, end, label=label, alignment_mode='contract')
                if span is None:
                    error_file.write(f"Error with annotation: {start}-{end} for text: {text}\n")
                else:
                    ents.append(span)

            try:
                doc.ents = ents
                db.add(doc)
                count += 1
            except Exception as e:
                error_file.write(f"Failed to add doc for text: {text} with error: {e}\n")

    print(f"Added {count} docs.")
    return db
```

```python
structured_data = cv_data[1]["structured_data"]
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(structured_data, test_size=0.3)
```

```python
len(train), len(test)
```

```python
file = open('/content/error_file.txt', 'w')

db = get_spacy_doc(file, train)
db.to_disk('/content/train_data.spacy')

db = get_spacy_doc(file, test)
db.to_disk('/content/test_data.spacy')

file.close()
```
